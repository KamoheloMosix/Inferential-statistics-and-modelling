## Why Remove dupes
- maintain accurate, consistent datasets
- avoid producing misleading statistics

```python
import numpy as np, pandas as pd
from pandas import Series, DataFrame

```

## REmoving duplicates breva

```python
df = DataFrame({'column 1': [1, 1, 2, 2, 3, 3, 3],
'column 2':['a', 'a', 'b', 'b', 'c', 'c', 'c'],
'column 3':['A', 'A', 'B', 'B', 'C', 'C', 'C']})
df
```

```python
# use duplicate method which shows us the duplicates
df.duplicated()
```

```python
# drop duplicates in the df
df.drop_duplicates()
```

```python
df = DataFrame({'column 1': [1, 1, 2, 2, 3, 3, 3],
'column 2':['a', 'a', 'b', 'b', 'c', 'c', 'c'],
'column 3':['A', 'A', 'B', 'B', 'C', 'C', 'D']})
df
```
