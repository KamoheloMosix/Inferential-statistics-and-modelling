```python
import pandas as pd
import numpy as np

from pandas import DataFrame
```

```python
numsdf = DataFrame(np.arange(0, 90, 3).reshape(10, 3), index=['r1', 'r2','r3','r4','r5','r6','r7','r8','r9','r10'],columns=['col_one','col_two','col_three'])
```

```python
numsdf
```

```python
# simple indexing, using iloc
numsdf.iloc[0,] # first 3 entries in r1

numsdf.iloc[1,1]

numsdf.iloc[5,2] = 17
```

```python
# get multiple rows and colums
# iloc[ rows, columns]
numsdf.iloc[[0,1,2], [0,1]]
```

```python
# boolean indexing using comparison operators, basica math operations e.g > , == etc
mask = numsdf > 45
mask

# can also be used to retrieve a set of values, the same way we use it in R
numsdf[mask]
```

```python
# used to replace certain nums
numsdf[mask] = 1
numsdf
```

```python
# now we are slciing buddy boy
#[startind:endindex], exclusive though
#[rows:rows, cols:cols]
numsdf.iloc[2:6, 1:3]
```
