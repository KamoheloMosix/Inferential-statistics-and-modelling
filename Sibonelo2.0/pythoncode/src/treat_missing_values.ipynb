## Missing values

- Approximating the missing values may be better than dropping entire rows or columns of missing information using:

1. *mean* approximation - from the values that we have we can get the mean and use that to approximate the missing values

### what this lesson covers
1. Discovering missing values
2. Fill in missing values
3. Counting missing values
4. Filtering out missing values

 here we learn how to treat missing values
 dataset with values 0, 99, or 9999 are considered as missing values


```python
import numpy as np
import pandas as pd
from pandas import DataFrame

```

#### key methods for missing vals
- fillna()
- replace()
- interpolate() - uses some kind of interpolation technique to fill-values

```python
# create dataset of people
data = {'Names': ['Sibonelo', 'Trust', 'Kamohelo', 'Rick', 'Morty', 'Goku', "Android 18"],
'Age': [20, 20, 19, 70, 14, 35, 30],
'Gender':["Male", 'Male', 'Male', 'Male', "Male", "Male", 'Female'],
'Rank':[2, 3, 1, 4, 5, 6, 7]}


```

```python
# create dataframe
rank_df = pd.DataFrame(data)
rank_df
```

```python
rank_df.iloc[2:5, 1] = np.nan  # set entries in first column, to nan
rank_df.iloc[3,:] = np.nan
rank_df.iloc[3:6, 3] = np.nan

# print dataFRame with missing values
rank_df
```

##### functions to detect missing values
1. pd.isnull() -> return true for missing values in df
2. pd.notnull() -> true for values that are not misiing

```python
rank_df.isnull()
```

```python
rank_df.notnull()
```

```python
# apply boolean masking to show rows where values in age column are missing
missing_df = pd.isnull(rank_df['Age'])
rank_df[missing_df]
```

```python
# use fill-in functions
rank_df.fillna(9)
```

```python
# let's try to fill value with value in prev index
# forward-fill
rank_df.ffill()

# fill in with the next entry
# backward-fill
rank_df.bfill()
```

```python
# use linear interpolation, treats values as if they're equally spaces
# rank_df.interpolate(method = 'linear')

# did not work becuase I have some non-number values
```

```python
# use dropna() to remove all rows with na-values
rank_df.dropna()
```

```python
# drop rows with only missing values
rank_df.dropna(how='all')

# axis, what dimension do we wanna apply this accross?
# default is 0
# axis = 0 ---> ROWS
# axis = 1 ---> COLUMNS
```
